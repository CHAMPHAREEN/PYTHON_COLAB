```python
# PLOT POINTS 
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([0,6])
ypoints = np.array([100,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
#PLOT GRAPH
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([3,7,10,12])
plt.plot(ypoints, marker = 'o')
plt.show()
plt.plot(ypoints, marker = '*')
plt.show()
```
